### Start stations histogram

```python
p = figure(x_range = topstartnames)
standardize(p)
standardizevbar(p)

x=topstartnames
temp = data[data.startstationid.isin(topstart)]
temp = temp.groupby('startstationid').count().max(1).values
temp = np.sort(temp)
temp = temp[::-1] # to sort in descending order
top = temp
width = 0.85
bottom = 0
p.vbar(x=x,top=top,width=width,bottom=bottom,color='skyblue');

pcts = top/data.shape[0]*100
pcts = np.round(pcts,2)
pctsStrings = []
for i in range(pcts.size):
    pctsStrings.append(str(pcts[i]) + "%")
source = ColumnDataSource(data=dict(x=x,
                                   top=top,
                                   pcts=pctsStrings))
labels = LabelSet(x='x',y='top',text='pcts',
                 level='glyph',x_offset=0,y_offset=5,
                 source=source,render_mode='canvas',
                 text_align='center',text_font_size='7pt',angle=-0.5)
p.add_layout(labels)
print(str(sum(pcts)) + '% of trips originated at these stations')

p.title.text = 'Trips Taken From Top 20 Origin Stations'
p.xaxis.axis_label = 'Station Name'
p.yaxis.axis_label = '# of trips'

p.left[0].formatter.use_scientific = False
```

```python
show(p)
```

### Most popular destination for each of the top 20 origin stations
- find the most used destinations from those top 20 origins
- how many of those destinations are in the overall top 20 destinations?

```python
temp = data[data.startstationid.isin(topstart)]
topdests = temp.groupby('endstationid').count().max(1).sort_values(ascending=False)[0:20]
```

```python
topdestnames = []
for idx,elem in enumerate(topdests.index):
    name = data[data.endstationid==elem].endstationname.unique()[0]
    print str(idx+1)+")"+name
    if np.intersect1d(name,topendnames).shape[0]==0:
        topdestnames.append(name)
    else :
        topdestnames.append('~ '+name)
print idx
topdestnames
```

```python
p = figure(x_range = topdestnames)
standardize(p)
standardizevbar(p)

x=topdestnames
top = topdests.values
width = 0.85
bottom = 0
p.vbar(x=x,top=top,width=width,bottom=bottom,color='skyblue');

pcts = top/data.shape[0]*100
pcts = np.round(pcts,2)
pctsStrings = []
for i in range(pcts.size):
    pctsStrings.append(str(pcts[i]) + "%")
source = ColumnDataSource(data=dict(x=x,
                                   top=top,
                                   pcts=pctsStrings))
labels = LabelSet(x='x',y='top',text='pcts',
                 level='glyph',x_offset=0,y_offset=5,
                 source=source,render_mode='canvas',
                 text_align='center',text_font_size='7pt',angle=-0.5)
p.add_layout(labels)
print(str(sum(pcts)) + '% of total trips ended at these stations')
print(str(sum(topdests.values/temp.shape[0])*100) + '% of trips taken from the top 20 stations ended at these stations')

p.title.text = 'Top 20 Destinations of Trips Taken From Top 20 Origin Stations'
p.xaxis.axis_label = 'Station Name'
p.yaxis.axis_label = '# of trips'

p.left[0].formatter.use_scientific = False
```
