### Most popular station for each type of customer
For this analysis, we will not distinguish between start/end...just if a station was used or not

** NOTE:  this means the same trip may be counted twice!  that's fine, just don't compare to other hists **

** KEY **

~ = in top 20 start stations

\* = in top 20 end stations

~\* = in both

```python
subscribers = data[data.usertype==1]
customers = data[data.usertype==2]
unknowns = data[data.usertype==0]

print subscribers.shape[0]
print customers.shape[0]
print unknowns.shape[0]
```

#### For subscribers

```python
ss = subscribers.groupby('startstationid').count().max(1).sort_values(ascending=False)[0:20]
se = subscribers.groupby('endstationid').count().max(1).sort_values(ascending=False)[0:20]

allPoss = np.union1d(ss.index,se.index)
totals = pd.Series(index=allPoss)
for idx,elem in enumerate(allPoss):
    try:
        a = ss.loc[elem]
    except:
        a=0
    try:
        b = se.loc[elem]
    except:
        b=0
    totals.loc[elem] = a+b
totals = totals.sort_values(ascending=False)
print totals.shape
totals
```

```python
toptotalnames = []
for idx,elem in enumerate(totals.index):
    name = data[data.startstationid==elem].startstationname.unique()[0]
    print str(idx+1)+")"+name
    if (np.intersect1d(name,topstartnames).shape[0]!=0 and np.intersect1d(name,topendnames).shape[0]!=0):
        toptotalnames.append('~* ' + name)
    elif np.intersect1d(name,topstartnames).shape[0]!=0:
        toptotalnames.append('~ ' + name)
    elif np.intersect1d(name,topendnames).shape[0]!=0:
        toptotalnames.append('* ' + name)
    else :
        toptotalnames.append(name)
print idx
toptotalnames
```

```python
p = figure(x_range = toptotalnames)
standardize(p)
standardizevbar(p)

x=toptotalnames
top = totals.values
width = 0.85
bottom = 0
p.vbar(x=x,top=top,width=width,bottom=bottom,color='skyblue');

p.title.text = 'Top 20 Stations Used By Subscribers'
p.xaxis.axis_label = 'Station Name'
p.yaxis.axis_label = '# of trips starting or ending'

p.left[0].formatter.use_scientific = False
```

### What is the ratio of start/stop for each of the stations?

```python
starts = data.startstationid.value_counts()
ends = data.endstationid.value_counts()
```

```python
ratios = pd.Series(index=starts.index)
ctr=0
for idx,elem in enumerate(starts.index):
    ratios.loc[elem] = starts.loc[elem]/ends.loc[elem]
#     if idx<100:
#         print starts.loc[elem]
#         print ends.loc[elem]
#         print '-----------------------------------'
    if starts.loc[elem]==ends.loc[elem]:
        ctr=ctr+1
ratios    
```
